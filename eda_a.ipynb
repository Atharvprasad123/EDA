# Exploratory Data Analysis (EDA)

This notebook performs exploratory data analysis (EDA) on the dataset, including basic data inspection, summary statistics, and visualizations.


## üì• Data Loading and Overview

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

goalscorers_df = pd.read_csv(r"C:\Users\atharv.prasad\Desktop\Atharv\datasets\goalscorers.csv")
results_df = pd.read_csv(r"C:\Users\atharv.prasad\Desktop\Atharv\datasets\results.csv")
shootouts_df = pd.read_csv(r"C:\Users\atharv.prasad\Desktop\Atharv\datasets\shootouts.csv")

# Convert 'date' columns to datetime format for all datasets
goalscorers_df['date'] = pd.to_datetime(goalscorers_df['date'])
results_df['date'] = pd.to_datetime(results_df['date'])
shootouts_df['date'] = pd.to_datetime(shootouts_df['date'])

# View the first few rows of the datasets
print(goalscorers_df.head())
print(results_df.head())
print(shootouts_df.head())

# Check for missing values
print(goalscorers_df.isnull().sum())
print(results_df.isnull().sum())
print(shootouts_df.isnull().sum())

# Get basic statistics for numeric columns
print(goalscorers_df.describe())
print(results_df.describe())
      
# Print the first few rows of the DataFrame   
print(goalscorers_df.head())
print(results_df.head())
print(shootouts_df.head())

# Display summary of the DataFrame
print(goalscorers_df.info())  
print(results_df.info())
print(shootouts_df.info())
 
# Check the columns
print(goalscorers_df.columns)
print(results_df.columns)
print(shootouts_df.columns)

print(goalscorers_df.shape)
print(results_df.shape)
print(shootouts_df.shape)

print(goalscorers_df.tail())
print(results_df.tail())
print(shootouts_df.tail())

print(goalscorers_df['home_team'].value_counts())
print(results_df['home_team'].value_counts())
print(shootouts_df['home_team'].value_counts())

print(goalscorers_df['away_team'].value_counts())
print(results_df['away_team'].value_counts())
print(shootouts_df['away_team'].value_counts())

print(goalscorers_df['team'].value_counts())
print(goalscorers_df['scorer'].value_counts())
print(goalscorers_df.loc[goalscorers_df['team'] == 'Portugal','scorer'])
print(goalscorers_df.loc[goalscorers_df['home_team'] == 'Portugal','scorer'])
print(goalscorers_df.loc[goalscorers_df['away_team'] == 'Portugal','scorer'])
print(goalscorers_df.loc[goalscorers_df['scorer'] == 'Cristiano Ronaldo','team'])
print(goalscorers_df.loc[goalscorers_df['team'] == 'Portugal','date'])
print(goalscorers_df.loc[goalscorers_df['home_team'] == 'Portugal','minute'].sum())
print(goalscorers_df.loc[goalscorers_df['away_team'] == 'Argentina','minute'].sum())
print(goalscorers_df.loc[goalscorers_df['team'] == 'Portugal','minute'].sum())
print(goalscorers_df.loc[goalscorers_df['scorer'] == 'Cristiano Ronaldo','minute'])
print(goalscorers_df.loc[goalscorers_df['scorer'] == 'Cristiano Ronaldo','penalty'])
print(goalscorers_df.loc[goalscorers_df['penalty'] == 'True','scorer'])

print(results_df['tournament'].value_counts())
print(results_df['country'].value_counts())
print(results_df['city'].value_counts())
print(results_df['home_score'].value_counts())
print(results_df['away_score'].value_counts())
print(results_df['neutral'].value_counts())
print(results_df.loc[results_df['home_team'] == 'Portugal','home_score'])
print(results_df.loc[results_df['home_team'] == 'Portugal','away_score'])
print(results_df.loc[results_df['tournament'] == 'FIFA World Cup','home_score'])
print(results_df.loc[results_df['home_team'] == 'Portugal','away_team'])

print(shootouts_df['winner'].value_counts())
print(shootouts_df.loc[shootouts_df['winner'] == 'India','date'])
print(shootouts_df.loc[shootouts_df['winner'] == 'India','away_team'])
print(shootouts_df.loc[shootouts_df['winner'] == 'India','home_team'])
print(shootouts_df.loc[shootouts_df['winner'] == 'India','first_shooter'])
print(shootouts_df.loc[shootouts_df['winner'] == 'Argentina','date'])
print(shootouts_df.loc[shootouts_df['winner'] == 'Argentina','away_team'])
print(shootouts_df.loc[shootouts_df['winner'] == 'Argentina','home_team'])
print(shootouts_df.loc[shootouts_df['winner'] == 'Argentina','first_shooter'])

# Distribution of goals by minute
plt.figure(figsize=(12, 7))
sns.histplot(goalscorers_df['minute'].dropna(), bins=60, kde=True, color="blue")
plt.title("Distribution of Goals by Minute", fontsize=18)
plt.xlabel("Minute", fontsize=14)
plt.ylabel("Number of Goals", fontsize=14)
plt.grid(True)
plt.tight_layout()
plt.show()

# Top 10 tournaments by number of matches
plt.figure(figsize=(14, 7))
tournament_counts = results_df['tournament'].value_counts().head(10)
sns.barplot(x=tournament_counts.index, y=tournament_counts.values, palette="viridis")
plt.title("Top 10 Tournaments by Number of Matches", fontsize=18)
plt.xlabel("Tournament", fontsize=14)
plt.ylabel("Number of Matches", fontsize=14)
plt.xticks(rotation=45, ha="right", fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()

# Top 10 goal scorers
plt.figure(figsize=(12, 7))
top_scorers = goalscorers_df['scorer'].value_counts().head(10)
sns.barplot(x=top_scorers.values, y=top_scorers.index, palette="Blues_d")
plt.title("Top 10 Goal Scorers", fontsize=18)
plt.xlabel("Number of Goals", fontsize=14)
plt.ylabel("Player", fontsize=14)
plt.grid(True)
plt.tight_layout()
plt.show()

# Filter for penalty goals
penalty_goals = goalscorers_df[goalscorers_df['penalty'] == True]
top_penalty_scorers = penalty_goals['scorer'].value_counts().head(10)

plt.figure(figsize=(12, 7))
sns.barplot(x=top_penalty_scorers.values, y=top_penalty_scorers.index, palette="Reds_r")
plt.title("Top 10 Penalty Goal Scorers", fontsize=18)
plt.xlabel("Number of Penalty Goals", fontsize=14)
plt.ylabel("Player", fontsize=14)
plt.grid(True)
plt.tight_layout()
plt.show()

# Filter for own goals
own_goals = goalscorers_df[goalscorers_df['own_goal'] == True]
top_own_goal_scorers = own_goals['scorer'].value_counts().head(10)

plt.figure(figsize=(12, 7))
sns.barplot(x=top_own_goal_scorers.values, y=top_own_goal_scorers.index, palette="Oranges_r")
plt.title("Top 10 Players by Own Goals", fontsize=18)
plt.xlabel("Number of Own Goals", fontsize=14)
plt.ylabel("Player", fontsize=14)
plt.grid(True)
plt.tight_layout()
plt.show()

# Filter Cristiano Ronaldo's goals
player = "Cristiano Ronaldo"
cristiano_goals = goalscorers_df[goalscorers_df['scorer'] == player]

# Function to plot bar charts
def plot_bar(data, x_col, y_col, title, xlabel, ylabel, palette="viridis"):
    plt.figure(figsize=(12, 7))
    sns.barplot(x=x_col, y=y_col, data=data, palette=palette)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Plot Goals Over Time
def plot_goals_over_time(goals_df, title, xlabel, ylabel, color="blue"):
    plt.figure(figsize=(12, 7))
    sns.histplot(goals_df['date'].dt.year, bins=30, color=color)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.grid(True)
    plt.tight_layout()
    plt.show()

plot_goals_over_time(cristiano_goals, f"Goals Over Time for {player}", "Year", "Number of Goals")

# Penalty Goals Over Time
penalty_goals = cristiano_goals[cristiano_goals['penalty'] == True]
plot_goals_over_time(penalty_goals, f"Penalty Goals Over Time for {player}", "Year", "Number of Penalty Goals", color="orange")

# Yearly Goals Count
yearly_goals = cristiano_goals.groupby(cristiano_goals['date'].dt.year)['scorer'].count()
plt.figure(figsize=(12, 7))
yearly_goals.plot(kind='line', marker='o', color="green")
plt.title(f"Goals Scored by {player} Each Year")
plt.xlabel("Year")
plt.ylabel("Number of Goals")
plt.grid(True)
plt.tight_layout()
plt.show()

# Goals by Age
birth_date = pd.to_datetime('1985-02-05')
cristiano_goals['age'] = (cristiano_goals['date'] - birth_date).dt.days // 365

plt.figure(figsize=(12, 7))
sns.histplot(cristiano_goals['age'], bins=20, kde=True, color="green")
plt.title(f"Goals by Age for {player}")
plt.xlabel("Age")
plt.ylabel("Number of Goals")
plt.tight_layout()
plt.show()

# Goals per Season
plt.figure(figsize=(12, 7))
seasonal_goals = cristiano_goals.groupby(cristiano_goals['date'].dt.year)['scorer'].count()
seasonal_goals.plot(kind='bar', color='skyblue')
plt.title("Goals per Season")
plt.xlabel("Season")
plt.ylabel("Goals")
plt.tight_layout()
plt.show()

# Filter own goals for Cristiano Ronaldo
own_goals = cristiano_goals[cristiano_goals['own_goal'] == True]

# Plot number of own goals per year (if any exist)
if len(own_goals) > 0:
    plt.figure(figsize=(12, 7))
    sns.histplot(own_goals['date'].dt.year, bins=30, color="red")
    plt.title(f"Own Goals Over Time for {player}", fontsize=18)
    plt.xlabel("Year", fontsize=14)
    plt.ylabel("Number of Own Goals", fontsize=14)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
else:
    print(f"{player} has no own goals recorded.")
    
# Count penalty vs non-penalty goals
penalty_goals = cristiano_goals[cristiano_goals['penalty'] == True].shape[0]
non_penalty_goals = cristiano_goals[cristiano_goals['penalty'] == False].shape[0]

# Pie chart for penalty vs non-penalty goals
plt.figure(figsize=(7, 7))
labels = ['Penalty Goals', 'Non-Penalty Goals']
sizes = [penalty_goals, non_penalty_goals]
colors = ['#FF9999', '#66B3FF']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90, explode=(0.1, 0))
plt.title(f"Penalty vs Non-Penalty Goals for {player}", fontsize=18)
plt.tight_layout()
plt.show()

# Filter for hat-tricks (assuming hat-tricks are defined as games with 3 or more goals)
hat_tricks = cristiano_goals.groupby(['date']).size().reset_index(name='goal_count')
hat_tricks = hat_tricks[hat_tricks['goal_count'] >= 3]

# Plot hat-tricks over time
plt.figure(figsize=(12, 7))
sns.histplot(hat_tricks['date'].dt.year, bins=30, color="red")
plt.title(f"Hat-Tricks Over Time for {player}", fontsize=18)
plt.xlabel("Year", fontsize=14)
plt.ylabel("Number of Hat-Tricks", fontsize=14)
plt.grid(True)
plt.tight_layout()
plt.show()

# Calculate goals per match
# Assuming each row in goalscorers_df represents a goal, so we need to aggregate by match
matches = cristiano_goals.groupby('date').size().reset_index(name='goals')
matches['match'] = matches.index + 1  # Create a match number column

# Plot Goals per Match
plt.figure(figsize=(12, 7))
sns.lineplot(x='match', y='goals', data=matches, marker='o', color='blue')
plt.title(f"Goals per Match for {player}", fontsize=18)
plt.xlabel("Match Number", fontsize=14)
plt.ylabel("Number of Goals", fontsize=14)
plt.grid(True)
plt.tight_layout()
plt.show()

# Filter Lionel Messi's goals
player = "Lionel Messi"
Messi_goals = goalscorers_df[goalscorers_df['scorer'] == player]

# Plot Goals Over Time
plot_goals_over_time(Messi_goals, f"Goals Over Time for {player}", "Year", "Number of Goals")

# Penalty Goals Over Time
penalty_goals = Messi_goals[Messi_goals['penalty'] == True]
plot_goals_over_time(penalty_goals, f"Penalty Goals Over Time for {player}", "Year", "Number of Penalty Goals", color="orange")

# Yearly Goals Count
yearly_goals = Messi_goals.groupby(Messi_goals['date'].dt.year)['scorer'].count()
plt.figure(figsize=(12, 7))
yearly_goals.plot(kind='line', marker='o', color="green")
plt.title(f"Goals Scored by {player} Each Year")
plt.xlabel("Year")
plt.ylabel("Number of Goals")
plt.grid(True)
plt.tight_layout()
plt.show()

# Goals by Age
birth_date = pd.to_datetime('1987-06-24')
Messi_goals['age'] = (Messi_goals['date'] - birth_date).dt.days // 365

plt.figure(figsize=(12, 7))
sns.histplot(Messi_goals['age'], bins=20, kde=True, color="green")
plt.title(f"Goals by Age for {player}")
plt.xlabel("Age")
plt.ylabel("Number of Goals")
plt.tight_layout()
plt.show()

# Goals per Season
plt.figure(figsize=(12, 7))
seasonal_goals = Messi_goals.groupby(Messi_goals['date'].dt.year)['scorer'].count()
seasonal_goals.plot(kind='bar', color='skyblue')
plt.title("Goals per Season")
plt.xlabel("Season")
plt.ylabel("Goals")
plt.tight_layout()
plt.show()

# Filter own goals for Lionel Messi
own_goals = Messi_goals[Messi_goals['own_goal'] == True]

# Plot number of own goals per year (if any exist)
if len(own_goals) > 0:
    plt.figure(figsize=(12, 7))
    sns.histplot(own_goals['date'].dt.year, bins=30, color="red")
    plt.title(f"Own Goals Over Time for {player}", fontsize=18)
    plt.xlabel("Year", fontsize=14)
    plt.ylabel("Number of Own Goals", fontsize=14)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
else:
    print(f"{player} has no own goals recorded.")

# Count penalty vs non-penalty goals
penalty_goals = Messi_goals[Messi_goals['penalty'] == True].shape[0]
non_penalty_goals = Messi_goals[Messi_goals['penalty'] == False].shape[0]

# Pie chart for penalty vs non-penalty goals
plt.figure(figsize=(7, 7))
labels = ['Penalty Goals', 'Non-Penalty Goals']
sizes = [penalty_goals, non_penalty_goals]
colors = ['#FF9999', '#66B3FF']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90, explode=(0.1, 0))
plt.title(f"Penalty vs Non-Penalty Goals for {player}", fontsize=18)
plt.tight_layout()
plt.show()

# Filter for hat-tricks (assuming hat-tricks are defined as games with 3 or more goals)
hat_tricks = Messi_goals.groupby(['date']).size().reset_index(name='goal_count')
hat_tricks = hat_tricks[hat_tricks['goal_count'] >= 3]

# Plot hat-tricks over time
plt.figure(figsize=(12, 7))
sns.histplot(hat_tricks['date'].dt.year, bins=30, color="red")
plt.title(f"Hat-Tricks Over Time for {player}", fontsize=18)
plt.xlabel("Year", fontsize=14)
plt.ylabel("Number of Hat-Tricks", fontsize=14)
plt.grid(True)
plt.tight_layout()
plt.show()

# Calculate goals per match
matches = Messi_goals.groupby('date').size().reset_index(name='goals')
matches['match'] = matches.index + 1  # Create a match number column

# Plot Goals per Match
plt.figure(figsize=(12, 7))
sns.lineplot(x='match', y='goals', data=matches, marker='o', color='blue')
plt.title(f"Goals per Match for {player}", fontsize=18)
plt.xlabel("Match Number", fontsize=14)
plt.ylabel("Number of Goals", fontsize=14)
plt.grid(True)
plt.tight_layout()
plt.show()

# Define additional top players
players = ["Cristiano Ronaldo", "Lionel Messi", "Neymar", "Robert Lewandowski", "Karim Benzema", "Romelu Lukaku", "Harry Kane"]

# Define birthdates for players (replace with actual birthdates if available)
birthdates = {
    "Cristiano Ronaldo": pd.to_datetime('1985-02-05'),
    "Lionel Messi": pd.to_datetime('1987-06-24'),
    "Neymar": pd.to_datetime('1992-02-05'),
    "Kylian Mbappe": pd.to_datetime('1998-12-20'),
    "Robert Lewandowski": pd.to_datetime('1988-08-21'),
    "Karim Benzema": pd.to_datetime('1987-12-19'),
    "Romelu Lukaku": pd.to_datetime('1993-05-13'),
    "Harry Kane": pd.to_datetime('1993-07-28'),
    "Luis Suarez": pd.to_datetime('1987-01-24')
}

# Filter data for selected players
players_df = goalscorers_df[goalscorers_df['scorer'].isin(players)]

# Calculate age at the time of each goal
players_df['birth_date'] = players_df['scorer'].map(birthdates)
players_df['age'] = (players_df['date'] - players_df['birth_date']).dt.days // 365

# Plot age distribution for each player
plt.figure(figsize=(14, 8))
for player in players:
    sns.histplot(players_df[players_df['scorer'] == player]['age'], bins=20, kde=True, label=player, alpha=0.6)

plt.title('Goals Scored by Age - Comparison of Selected Players', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Number of Goals', fontsize=14)
plt.legend(title='Player')
plt.grid(True)
plt.tight_layout()
plt.show()

# Goals per Season for all players
plt.figure(figsize=(14, 8))
seasonal_goals_all = players_df.groupby(['scorer', players_df['date'].dt.year])['scorer'].count().unstack()
seasonal_goals_all.plot(kind='bar', stacked=True, figsize=(14, 8), colormap='tab10')
plt.title("Goals per Season for Selected Players", fontsize=18)
plt.xlabel("Season", fontsize=14)
plt.ylabel("Number of Goals", fontsize=14)
plt.legend(title='Player')
plt.grid(True)
plt.tight_layout()
plt.show()


# Plotting goals by minute for all selected players
plt.figure(figsize=(14, 8))
for player in players:
    sns.histplot(players_df[players_df['scorer'] == player]['minute'].dropna(), bins=60, kde=True, label=player, alpha=0.6)

plt.title('Distribution of Goals by Minute - Comparison of Selected Players', fontsize=16)
plt.xlabel('Minute', fontsize=14)
plt.ylabel('Number of Goals', fontsize=14)
plt.legend(title='Player')
plt.grid(True)
plt.tight_layout()
plt.show()

# Penalty vs Non-Penalty Goals for all players
penalty_vs_non_penalty = players_df.groupby(['scorer', 'penalty'])['scorer'].count().unstack().fillna(0)
penalty_vs_non_penalty.columns = ['Non-Penalty Goals', 'Penalty Goals']

plt.figure(figsize=(14, 8))
penalty_vs_non_penalty.plot(kind='bar', figsize=(14, 8), colormap='Set1')
plt.title("Penalty vs Non-Penalty Goals for Selected Players", fontsize=18)
plt.xlabel("Player", fontsize=14)
plt.ylabel("Number of Goals", fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.legend(title='Goal Type')
plt.grid(True)
plt.tight_layout()
plt.show()

# Goals by Age for each player (Separate charts)
for player in players:
    player_goals = players_df[players_df['scorer'] == player]
    
    plt.figure(figsize=(12, 7))
    sns.histplot(player_goals['age'], bins=20, kde=True, color="green")
    plt.title(f"Goals by Age for {player}")
    plt.xlabel("Age")
    plt.ylabel("Number of Goals")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Goals per Match for each player (Separate charts)
for player in players:
    player_goals = players_df[players_df['scorer'] == player]
    matches = player_goals.groupby('date').size().reset_index(name='goals')
    matches['match'] = matches.index + 1  # Create a match number column

    plt.figure(figsize=(12, 7))
    sns.lineplot(x='match', y='goals', data=matches, marker='o', color='blue')
    plt.title(f"Goals per Match for {player}")
    plt.xlabel("Match Number")
    plt.ylabel("Number of Goals")
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
